```python
import numpy as np
import matplotlib.pyplot as plt
```

## Ice Cavity

```python
def process(IS_name, mode, max_time):

    Stucked = np.load(f'processed_dataGL/{IS_name}_{mode}_Stucked.npy') 
    IS = np.load(f'processed_dataGL/{IS_name}_{mode}_IS.npy') 
    
    n = int(len(IS)/366)
    
    resIn = []
    resOut = []
    
    for p in range(0, len(IS), n):
        
        sOut = IS[p:p+n]
        sStucked = Stucked[p:p+n]
        
    
        ## Remove Stucked parcels
        sOut[sStucked == 1] = np.nan


        ## Remove particles slower than X
        sOut[sOut > max_time] = np.nan


        ## Calculate total particles
        total_particles = len(sStucked) - np.count_nonzero(sStucked == 1)
    
        ## Get num of particles per region
        c1 = np.count_nonzero(~np.isnan(sOut))
        c2 = total_particles - c1
        
        resOut.append(c1/total_particles*100)
        resIn.append(c2/total_particles*100)
    
    
    print(f"{np.mean(resOut)} + {np.std(resOut)}")
    print(f"{np.mean(resIn)} + {np.std(resIn)}")
    
    return (np.mean(resOut), np.mean(resIn))
```

```python
IS_name = 'pineisland'
mode = '0d'
max_time = 30

sizes = process(IS_name = IS_name,
                         mode = mode,
                         max_time=max_time)

fig, ax = plt.subplots()


wedges, labels = ax.pie([sizes[0], sizes[1]],
      colors=['blue', 'gray'],
       textprops = {"fontsize":25})

for lab in labels:
    lab.set_fontsize(15)
    
plt.tight_layout()

plt.savefig(f'img_composition/Cavity_{IS_name}_{max_time:03d}.png', dpi=300)
```

## Continental Shelf

```python
def process(IS_name, mode, max_time):
    
    ## Load data
    East = np.load(f'processed_dataGL/{IS_name}_{mode}_Eastern.npy') 
    West = np.load(f'processed_dataGL/{IS_name}_{mode}_Western.npy') 
    CS = np.load(f'processed_dataGL/{IS_name}_{mode}_CS.npy') 
    Stucked = np.load(f'processed_dataGL/{IS_name}_{mode}_Stucked.npy') 
    IS = np.load(f'processed_dataGL/{IS_name}_{mode}_IS.npy') 
    
    n = int(len(IS)/366)
    
    res_SE = []
    res_SW = []
    res_CS = []
    res_OS = []
    
    for p in range(0, len(IS), n):
        sE = East[p:p+n]
        sW = West[p:p+n]
        sIS = IS[p:p+n]
        sCS = CS[p:p+n]
        sStucked = Stucked[p:p+n]
        
    
    
        ## Remove Stucked parcels
        sE[sStucked == 1] = np.nan
        sW[sStucked == 1] = np.nan
        sIS[sStucked == 1] = np.nan
        sCS[sStucked == 1] = np.nan

        ## Remove particles slower than X
        sE[sE > max_time] = np.nan
        sW[sW > max_time] = np.nan
        sCS[sCS > max_time] = np.nan
    
        ## Remove particles from East and West if they made it off shelf
        sE[~np.isnan(sCS)] = np.nan
        sW[~np.isnan(sCS)] = np.nan

        ## Calculate total particles
        total_particles = len(sStucked) - np.count_nonzero(sStucked == 1)
    
        ## Get num of particles per region
        c1 = np.count_nonzero(~np.isnan(sE))
        c2 = np.count_nonzero(~np.isnan(sW))
        c3 = np.count_nonzero(~np.isnan(sCS))
        c4 = total_particles - (c1 + c2 + c3)
        
        res_SE.append(c1/total_particles*100)
        res_SW.append(c2/total_particles*100)
        res_CS.append(c4/total_particles*100)
        res_OS.append(c3/total_particles*100)
    
    
    print(f"{np.mean(res_SE)} + {np.std(res_SE)}")
    print(f"{np.mean(res_SW)} + {np.std(res_SW)}")
    print(f"{np.mean(res_CS)} + {np.std(res_CS)}")
    print(f"{np.mean(res_OS)} + {np.std(res_OS)}")
    
    return (np.mean(res_OS), np.mean(res_SE), np.mean(res_CS))
```

```python
IS_name = 'pineisland'
mode = '0d'
max_time = 365

sizes = process(IS_name = IS_name,
                         mode = mode,
                         max_time=max_time)

fig, ax = plt.subplots()


wedges, labels = ax.pie([sizes[0], sizes[1], sizes[2]],
      colors=['red', 'cyan', 'gray'],
       textprops = {"fontsize":25})

for lab in labels:
    lab.set_fontsize(15)
    
plt.tight_layout()

plt.savefig(f'img_composition/Shelf_{IS_name}_{max_time:03d}.png', dpi=300)
```
